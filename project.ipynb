```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df.isnull().sum()
```

```python
df = df.drop('Cabin', axis =1)
```

```python
df.head()
```

```python
df.Sex.value_counts()
```

```python
df['Sex'] = df['Sex'].replace({'male' : 1 , 'female' : 0})
```

```python
df.head()
```

```python
df = df.drop(['Name', 'Sex', 'Ticket', 'Embarked'], axis=1)
```

```python
df.head()
```

```python
df['Survived_'] = df['Survived']
```

```python
df.head()
```

```python
df = df.drop('Survived' , axis =1 )
```

```python
df.head()
```

```python
df.rename( columns= {'Survived_': 'Survived'})
```

```python
X = df.iloc[: , : -1]
X
```

```python
Y = df.iloc[: , -1]
Y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, f1_score,accuracy_score

confusion_matrix(Y_test, y_pred)
```

```python
score = accuracy_score(Y_test, y_pred)
score
```

```python
from sklearn.model_selection import GridSearchCV
modellr = LogisticRegression()

params = {
    'penalty' : ['l1' , 'l2', 'elasticnet'],
    'C' :[0.01, 0.1, 1, 10, 100],
    'solver' : ['lbfgs', 'liblinear', 'newton-cg', 'sag', 'saga'],
    'max_iter' : [10,100,200,400],
    'l1_ratio' : [0,1,0.5, 0.6]
}
```

```python
grid =GridSearchCV(estimator= modellr, param_grid=params, scoring='accuracy', cv = 10, n_jobs=-1)
grid.fit(X_train, Y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
matrix = pd.DataFrame(confusion_matrix(Y_test, y_pred), columns= ['Predicted Yes', 'Predicted Yes'], index = ['Actual Yes', 'Actual No'])
matrix
```

```python
f1= f1_score(Y_test, y_pred)
f1
```

```python
from sklearn.metrics import classification_report

report = classification_report(Y_test, y_pred)
```

```python
print(report)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model2 = KNeighborsClassifier(n_neighbors=10, algorithm= 'ball_tree')
model2.fit(X_train, Y_train)
y_pred2 = model2.predict(X_test)
print(accuracy_score(Y_test, y_pred2))
```

```python
from sklearn.svm import SVC

model3 = SVC()
model3.fit(X_train, Y_train)
y_pred3 = model3.predict(X_test)
print(accuracy_score(Y_test, y_pred3))
```

```python
from sklearn.tree import DecisionTreeClassifier

model4= DecisionTreeClassifier(max_depth=5)
model4.fit(X_train, Y_train)
y_pred4 = model4.predict(X_test)
print(accuracy_score(Y_test, y_pred4))
```

```python
from sklearn.ensemble import RandomForestClassifier

model5= RandomForestClassifier(max_depth=15, max_features=1, max_samples=0.7, n_estimators=400, n_jobs=-1, bootstrap=True)
model5.fit(X_train, Y_train)
y_pred5 = model5.predict(X_test)
print(accuracy_score(Y_test, y_pred5))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

model6= GradientBoostingClassifier()
model6.fit(X_train, Y_train)
y_pred6 = model6.predict(X_test)
print(accuracy_score(Y_test, y_pred6))
```
